```python
import numpy as np
import matplotlib.pyplot as plt
```

Dados da placa de Orifício

```python
def velocidade(a, b, rho, rhom, mu, dH, Dt, Or, p):

  Dt = Dt*10**-3 #Diâmetro da tubulação
  Or = Or*10**-3  #Diâmetro do orifício
  beta = Or/Dt

  Ar_Dt = np.pi*Dt**2/4
  Ar_Or = np.pi*Or**2/4

  r2 = (Ar_Or/Ar_Dt)**2

  #Perda de Carga na placa
  #dH = np.linspace(0.01, 0.96, 20)
  dP = rhom*g*dH

  #Velocidade Teórica
  Vt = np.sqrt(2*dP/(rho*(1-r2)))

  #Vazão mássica teórica
  mt = rho*Vt*Ar_Dt

  #Reynolds Máximo
  Re = rho*Vt*Dt/mu

  #Coeficiente de descarga
  C = 0.5959 + 0.0312*beta**2.1 - 0.184*beta**8 + 91.71*(beta**2.5)/(Re**0.75)

  #Coeficiente de vazão
  k = C/np.sqrt(1-(beta**4))

  #Vazão Real
  mr = k*Ar_Dt*np.sqrt(2*rho*dP)

  #Vazão volumétrica
  Q = mr/rho

  #velocidade Real
  vel = mr/(rho*Ar_Dt)

  #Velocidade na entrada
  VelEn = Ar_Dt*vel/(a*b)

  if p==1:
    plt.plot(dH,VelEn)
    plt.title("Velocidade na Entrada do Ciclone")
    plt.xlabel("Altura Manométrica (m)")
    plt.ylabel("Velocidade na Entrada do Ciclone (m/s)")
    plt.show()

    print("\nOs valores:")

    for i in range(len(dH)):
        print(f"Velocidade na Entrada do Ciclone é: {VelEn[i]:.4f} m/s para {dH[i]:.4f} m de altura.")


  return VelEn, Q
```

```python
import numpy as np
import matplotlib.pyplot as plt

def velocidade2(a, b, rho, rhom, mu, dH, Dt, Or, p):
    # Constante gravitacional (certifique-se de definir ou passar como argumento)
    g = 9.81

    # Conversão de mm para metros
    Dt = Dt * 10**-3 # Diâmetro da tubulação
    Or = Or * 10**-3 # Diâmetro do orifício

    # Razão de diâmetros (beta)
    beta = Or / Dt

    # Áreas das seções transversais
    Ar_Dt = np.pi * (Dt**2) / 4 # Área do tubo
    Ar_Or = np.pi * (Or**2) / 4 # Área do orifício

    # Diferença de pressão (Pa) com base na altura manométrica dH
    # Nota: Se for manômetro de coluna de líquido, a equação rigorosa é (rhom - rho)*g*dH
    dP = rhom * g * dH

    # Velocidade Teórica NO ORIFÍCIO
    Vt_orificio = np.sqrt((2 * dP) / (rho * (1 - beta**4)))

    # Reynolds na tubulação (baseado na velocidade teórica na tubulação para aproximar C)
    # Vt_tubo = Vt_orificio * (beta**2)
    V_tubo_teorica = Vt_orificio * (beta**2)
    Re = (rho * V_tubo_teorica * Dt) / mu

    # Coeficiente de descarga (C) - Equação informada
    C = 0.5959 + 0.0312 * (beta**2.1) - 0.184 * (beta**8) + 91.71 * (beta**2.5) / (Re**0.75)

    # Coeficiente de vazão (K)
    k = C / np.sqrt(1 - (beta**4))

    # Vazão Mássica Real (Usando K e a área do orifício Ar_Or)
    mr = k * Ar_Or * np.sqrt(2 * rho * dP)

    # Vazão volumétrica (m³/s)
    Q = mr / rho

    # Velocidade Real na TUBULAÇÃO (m/s)
    vel_tubo = Q / Ar_Dt

    # Velocidade na entrada do Ciclone (Continuidade: Q = V * A -> V = Q / A_ciclone)
    VelEn = Q / (a * b)

    if p == 1:
        plt.plot(dH, VelEn)
        plt.title("Velocidade na Entrada do Ciclone")
        plt.xlabel("Altura Manométrica (m)")
        plt.ylabel("Velocidade na Entrada do Ciclone (m/s)")
        plt.grid(True)
        plt.show()

        print("\nOs valores:")
        # Garantindo que funciona tanto para um único valor de dH quanto para arrays
        if isinstance(dH, (list, np.ndarray)):
            for i in range(len(dH)):
                print(f"Velocidade na Entrada do Ciclone é: {VelEn[i]:.4f} m/s para {dH[i]:.4f} m de altura.")
        else:
            print(f"Velocidade na Entrada do Ciclone é: {VelEn:.4f} m/s para {dH:.4f} m de altura.")

    return VelEn, Q
```
